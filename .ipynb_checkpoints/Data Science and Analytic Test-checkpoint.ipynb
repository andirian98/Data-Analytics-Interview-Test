```python
import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt

new_data = pd.read_csv('C:/Users/User/Documents/Andy Docs/shopee_sample_data (1).csv')
```

<h1> Task 1</h1>

```python
#removing special characters for every value in title, desc, link_ori, location
for i in new_data.index:
    try:
        new_data.loc[i,'title'] = new_data.loc[i,'title'].replace("| Shopee Malaysia","")
    except:
        #new_data.loc[i,'title'] = str(new_data.loc[i,'title']).replace("| Shopee Malaysia","")
        new_data.loc[i,'title'] = str(new_data.loc[i,'title']).replace("| Shopee Malaysia","")

for i in new_data.index:
    try:
        new_data.loc[i,'title'] = re.sub(r"[^a-zA-z0-9.,!?/:;\"\'\s]","",new_data.loc[i,'title'])
    except:
        new_data.loc[i,'title'] = new_data.loc[i,'title']
        
for i in new_data.index:
    try:
        new_data.loc[i,'desc'] = re.sub(r"[^a-zA-z0-9.,!?/:;\"\'\s]","",new_data.loc[i,'desc'])
    except:
        #new_data.loc[i,'title'] = str(new_data.loc[i,'title']).replace("| Shopee Malaysia","")
        new_data.loc[i,'desc'] = re.sub(r"[^a-zA-z0-9.,!?/:;\"\'\s]","",str(new_data.loc[i,'desc']))
                                        
for i in new_data.index:
    try:
        new_data.loc[i,'link_ori'] = re.sub(r"[^a-zA-z0-9.,!?/:;\"\'\s\-]","",new_data.loc[i,'link_ori'])
    except:
        #new_data.loc[i,'title'] = str(new_data.loc[i,'title']).replace("| Shopee Malaysia","")
        new_data.loc[i,'link_ori'] = re.sub(r"[^a-zA-z0-9.,!?/:;\"\'\s\-]","",str(new_data.loc[i,'link_ori']))

for i in new_data.index:
    try:
        new_data.loc[i,'location'] = re.sub(r"[^a-zA-z0-9.,!?/:;\"\'\s\-]","",new_data.loc[i,'location'])
    except:
        new_data.loc[i,'location'] = new_data.loc[i,'location']
                                        
#Before null value filled in
print(new_data.location.isna().sum())

new_data.location.fillna("N/A", inplace = True)

#After null value filled in
print(new_data.location.isna().sum())

print(new_data)
```

```python
#to convert integer to datetime for every w_date values
for i in new_data.index:
    try:
        new_data.loc[i,'w_date'] = pd.to_datetime(new_data.loc[i,'w_date'] , format='%Y%m%d')
    except:
        new_data.loc[i,'w_date'] = new_data.loc[i,'w_date']

#find and print out products listed during Nov 2020
date_strt, date_end = pd.to_datetime(20201101.0, format='%Y%m%d'),pd.to_datetime(20201129.0, format='%Y%m%d')

print("The listed product during November 2020")
for i in new_data.index:
    if new_data.loc[i,'w_date'] >= date_strt and new_data.loc[i,'w_date'] <= date_end:
        print(new_data.loc[i,'title'])
```
